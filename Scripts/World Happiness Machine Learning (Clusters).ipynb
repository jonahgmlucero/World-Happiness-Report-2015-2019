# World Happiness: Unsupervised Machine Learning (Clusters)

### Importing libraries and data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.cluster import KMeans # Here is where you import the k-means algorithm from scikit-learn.
import pylab as pl # PyLab is a convenience module that bulk imports matplotlib.
```

```python
# To keep matplotlib visuals to appear in the notebook 
%matplotlib inline
```

```python
#creating a path
path = r'/Users/jonahsnewmacbook/Desktop/Data Analytics/Data Immersion/Achievement 6/World Happiness'
```

```python
#import World Happiness Dataframe
WH = pd.read_csv(os.path.join(path, 'Data', 'WorldHappinesscleaned.csv'))
```

```python
# This data has been cleaned from nulls and duplicates
```

```python
WH.head()
```

```python
# drop columns: 'Country', 'Region', 'Happiness Category'
```

```python
#Drop 'Region', and 'Standard Error'
WH = WH.drop('Region', axis=1)
WH = WH.drop('Country', axis=1)
WH = WH.drop('Happiness Category', axis=1)
```

```python
WH.head()
```

### Applying the Elbow Technique

```python
num_cl = range(1, 10) # defines the range of potential clusters
kmeans = [KMeans(n_clusters=i) for i in num_cl] #defines k-means clusters in the range assigned above
```

```python
score = [kmeans[i].fit(WH).score(WH) for i in range(len(kmeans))]
#this will create a score that shows a rate of variation for the given cluster option

score
```

```python
# Plotting the elbox curve using PyLab

pl.plot(num_cl,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

There's a jump from two to three. Graphs plateaus after three. This means that THREE is the optimal count for the dataframe's clusters.

### Applying K-Means Clustering

```python
# Create the k-means object
kmeans = KMeans(n_clusters = 3)
```

```python
#Fit the k-means object to the data
kmeans.fit(WH)
```

```python
WH['clusters'] = kmeans.fit_predict(WH)
```

```python
WH.head()
```

```python
WH['clusters'].value_counts()
```

```python
# Plot the clusters for 'Happiness Score' and 'Economy (GDP per Capita)' variables

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=WH['Happiness Score'], y=WH['Economy (GDP per Capita)'], hue=kmeans.labels_, s=100) 

ax.grid(False) # This removes the grid from the background.
plt.xlabel('Happiness Score') # Label x-axis.
plt.ylabel('Economy (GDP per Capita)') # Label y-axis.
plt.show()
```

This result is to be expected. It confirms that the higher the Happiness Score, the higher the Economy (GDP per Capita)

### Running a Descriptive Statistics on the clusters

```python
WH.loc[WH['clusters'] == 2, 'cluster'] = 'dark purple'
WH.loc[WH['clusters'] == 1, 'cluster'] = 'purple'
WH.loc[WH['clusters'] == 0, 'cluster'] = 'pink'
```

```python
WH.groupby('cluster').agg({'Happiness Score':['mean', 'median'], 
                         'Economy (GDP per Capita)':['mean', 'median'],
                          'Family':['mean', 'median'],
                          'Health (Life Expectancy)':['mean', 'median'],
                          'Freedom':['mean', 'median'],
                          'Trust (Government Corruption)':['mean', 'median'],
                          'Generosity':['mean', 'median']})
```

Observation:
- Pink cluster have the lowest mean and median except for two variables: 'Trust (Government Corruption)' and 'Generosity'.
- Purple cluster have the highest mean and median overall.
